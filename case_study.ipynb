##1: Load SpaceX Launch Data from API

```python
import requests
import pandas as pd

# API endpoint for SpaceX lau
LAUNCHES_API_URL = "https://api.spacexdata.com/v4/launches"

# Fetch data from the API
response = requests.get(LAUNCHES_API_URL)
launches_data = response.json()

launches_df = pd.DataFrame(launches_data)[['name', 'date_utc', 'success', 'details', 'rocket']]

launches_df['date_utc'] = pd.to_datetime(launches_df['date_utc'])
launches_df['launch_year'] = launches_df['date_utc'].dt.year

display(launches_df.head())
display(launches_df.info())
```

## Step 2: Load SpaceX Rocket Data from API

Load Launch Data

```python
import requests
import pandas as pd

# Fetch launch data
launch_url = "https://api.spacexdata.com/v4/launches"
launch_data = requests.get(launch_url).json()

# Convert to DataFrame
launch_df = pd.DataFrame(launch_data)

# Select required columns
launch_df = launch_df[['name', 'date_utc', 'success', 'details', 'rocket']]

# Convert date and extract year
launch_df['date_utc'] = pd.to_datetime(launch_df['date_utc'])
launch_df['year'] = launch_df['date_utc'].dt.year

launch_df.head()
```

Load Rocket Metadata

```python
rocket_url = "https://api.spacexdata.com/v4/rockets"
rocket_data = requests.get(rocket_url).json()

rocket_df = pd.DataFrame(rocket_data)

rocket_df = rocket_df[['id', 'name', 'type', 'active', 'stages']]

rocket_df.head()
```

 Merge Data

```python
merged_df = launch_df.merge(rocket_df, left_on='rocket', right_on='id', how='left')

merged_df.head()
```

Add Simulated Country

```python
import numpy as np

countries = ['USA', 'Russia', 'India', 'China', 'France']
merged_df['country'] = np.random.choice(countries, size=len(merged_df))

merged_df.head()
```

Store in SQLite

```python
import sqlite3

conn = sqlite3.connect("spacex_launches.db")

merged_df.to_sql("launches", conn, if_exists="replace", index=False)

print("Data stored successfully!")
```

SQL Analysis Queries

```python

query1 = """
SELECT country, COUNT(*) as total_launches
FROM launches
GROUP BY country
ORDER BY total_launches DESC;
"""

pd.read_sql_query(query1, conn)
```

```python
query2 = """
SELECT year, COUNT(*) as total_launches
FROM launches
GROUP BY year
ORDER BY total_launches DESC
LIMIT 1;
"""

pd.read_sql_query(query2, conn)
```

```python
query3 = """
SELECT name_x, COUNT(*) as launch_count
FROM launches
GROUP BY name_x
ORDER BY launch_count DESC
LIMIT 5;
"""

pd.read_sql_query(query3, conn)
```
